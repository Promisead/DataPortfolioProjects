# Merge, Join and Concatenate in Pandas

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\IT SUPPORT M4YORS PC\Desktop\1. Tech Videos\DATA ANALYSIS\files_folder\LOTR.csv")
```

```python
df2 = pd.read_csv(r"C:\Users\IT SUPPORT M4YORS PC\Desktop\1. Tech Videos\DATA ANALYSIS\files_folder\LOTR 2.csv")
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how = 'inner')
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
#Default
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_LHS', rsuffix = '_RHS')
```

# Joining on FellowshipID ==> Numbers

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix ='_LHS', rsuffix='_RHS')
```

```python
df4
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix ='_LHS', rsuffix='_RHS', how='outer')
df4
```

# Concatenation

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join='inner')
```

```python
pd.concat([df1,df2], join='outer')
```

```python
pd.concat([df1,df2], join='outer',axis = 1)
```

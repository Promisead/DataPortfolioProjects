```python
df2 = df.set_index('Country')
```

```python
df2
```

```python
df2.filter(items = ['Continent','CC3'], axis=1)
```

```python
df2.filter(like = 'United', axis=0)
```

```python
df2.loc['United States']
```

```python
df[df['Rank'] < 10].sort_values(by = ['Continent','Country'], ascending = [False, True])
```
